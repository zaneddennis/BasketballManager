```python
df = pd.read_sql(
    """
SELECT *
FROM Players
    """,
    conn,
    index_col="ID"
)
df
```

```python
df_active = df.loc[~df.SchoolID.isna()]
df_active
```

```python
def investigate_attr(c, b=16):
    fig, axs = plt.subplots(1, 2, sharex=True, sharey=True, tight_layout=True)
    axs[0].hist(df[c], bins=b)
    axs[1].hist(df_active[c], bins=b)
    axs[0].set_title("Recruits (n=920)")
    axs[1].set_title("Rostered (n=552)")
    fig.suptitle(c)
    plt.xlim(0.0, 1.0)
    plt.show()
```

```python
df["PolishStrength"] = df.Strength / df.PotentialStrength
df["PolishAgility"] = df.Agility / df.PotentialAgility
df
```

```python
sns.barplot(df, x="Eligibility", y="PolishStrength", order=["FR", "SO", "JR", "SR"])
plt.ylim(0.0, 1.0)
plt.title("Polish - Strength")
```

```python
sns.barplot(df, x="Eligibility", y="PolishAgility", order=["FR", "SO", "JR", "SR"])
plt.ylim(0.0, 1.0)
plt.title("Polish - Agility")
```
