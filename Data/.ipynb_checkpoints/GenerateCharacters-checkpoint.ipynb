```python
import pandas as pd

import numpy as np
from scipy.stats import skewnorm
from sklearn.linear_model import LinearRegression
import seaborn as sns

import sqlite3
```

```python
NUM_COACHES = 100
NUM_PLAYERS = 900
```

```python
OUTPUT = False
```

## Characters

```python
dfs_firsts = {}

for year in range(1960, 2023 + 1):
    df = pd.read_csv("Names/yob{}.txt".format(year), header=None)
    df.columns = ["Name", "Sex", "Count"]
    df = df.loc[df.Sex == "M"]
    dfs_firsts[year] = df
```

```python
df_surnames = pd.read_csv("Names/Names_2010Census.csv")[["name", "count"]]
df_surnames = df_surnames.drop(162253)  # throw out "ALL OTHER NAMES"
```

```python
df_locations = pd.read_csv("locations.tsv", sep="\t")[["ID", "Population"]]
df_locations = df_locations.loc[df_locations.Population >= 1000]

df_schools = pd.read_csv("schools_filtered.tsv", sep="\t")
```

```python
def gen_first(year):
    df_names = dfs_firsts[year]
    return df_names.sample(weights=df_names.Count).Name.iloc[0]

def gen_last(_=None):
    return df_surnames.sample(weights=df_surnames["count"]).name.iloc[0].capitalize()
```

```python
def gen_hometown(_=None):
    return df_locations.sample(weights=df_locations.Population).ID.iloc[0]

def gen_am(_=None):
    return df_schools.Abbreviation.sample().iloc[0]
```

```python
df = pd.DataFrame(
    [
        [int(np.random.uniform(1960, 1990))] for i in range(NUM_COACHES)
    ] + [
        [int(np.random.uniform(2002, 2006+1))] for i in range(NUM_PLAYERS)
    ],
    columns=["Birth"]
)

df["First"] = df.Birth.map(gen_first)
df["Last"] = df.index.map(gen_last)

df["Hometown"] = df.index.map(gen_hometown)
df.loc[df.index < NUM_COACHES, "AlmaMater"] = pd.Series(range(NUM_COACHES)).map(gen_am)
df["AlmaMater"] = df["AlmaMater"].fillna("")

df.loc[df.index < NUM_COACHES, "CoachID"] = range(1, NUM_COACHES+1)
df.loc[df.index >= NUM_COACHES, "PlayerID"] = range(1, NUM_PLAYERS+1)
df["CoachID"] = df["CoachID"].fillna(0).astype(int)
df["PlayerID"] = df["PlayerID"].fillna(0).astype(int)
df["ID"] = df.index + 1

df = df[["ID", "First", "Last", "Birth", "Hometown", "AlmaMater", "CoachID", "PlayerID"]]

df
```

## Coaches

## Players

```python
h1 = [73, 73, 74, 75, 75, 76, 77, 79, 80, 80, 81, 81, 82, 84]
w1 = [185, 180, 205, 180, 195, 195, 250, 200, 195, 245, 225, 270, 245, 215]

h2 = [71, 72, 75, 76, 77, 77, 78, 79, 80, 80, 81, 81, 84]
w2 = [175, 225, 215, 190, 200, 230, 205, 230, 225, 240, 205, 240, 225]

h3 = [74, 74, 76, 76, 77, 77, 78, 79, 80, 80, 82, 82, ]
w3 = [190, 180, 185, 190, 210, 195, 195, 220, 245, 230, 230, 235]

h = h1 + h2 + h3
w = w1 + w2 + w3

linreg_weight = LinearRegression().fit(
    np.array(h).reshape(-1, 1),
    np.array(w).reshape(-1, 1)
)
print(linreg_weight.coef_)
print(linreg_weight.intercept_)

sns.scatterplot(x=h, y=w)
```

```python
def gen_height():
    return round(skewnorm.rvs(a=0.5, loc=76, scale=3.6))

def gen_weight(height):
    w_raw = linreg_weight.predict([[height]])[0, 0]
    w_adj = w_raw + np.random.uniform(-3, 3) * 5.0
    return round(w_adj / 5.0) * 5

def get_height_adj(height):
    mean_height = np.mean(h)
    std_height = np.std(h)

    diff = height - mean_height
    return diff / std_height

def get_weight_adj(weight):
    mean_weight = np.mean(w)
    std_weight = np.std(w)

    diff = weight - mean_weight
    return diff / std_weight
```

```python
def gen_technical(height):
    return np.random.normal(10 - get_height_adj(height), 3)

def gen_mental(_):
    return np.random.normal(10, 3)
```

```python
def gen_agility(weight):
    return int(np.random.normal(10 - 2 * get_weight_adj(weight), 3))

def gen_strength(weight):
    return int(np.random.normal(10 + 3 * get_weight_adj(weight), 2))

def gen_vertical(height):
    return int(
        np.random.normal(10 + 3.5 * get_height_adj(height), 2)
    )
```

```python
def gen_ball_handling(row):
    height_adj = get_height_adj(row["Height"])
    tech = row["_technical"]
    
    return int(
        np.random.normal(tech - 2.5 * height_adj, 1.5)
    )

def gen_finishing(tech):
    return int(
        np.random.normal(np.mean([10, tech]), 2.5)
    )

def gen_shooting(tech):
    return int(
        np.random.normal(tech, 1.5)
    )

def gen_rebounding(_):
    return int(
        np.random.normal(10, 3)
    )

def gen_per_def(_):
    return int(
        np.random.normal(10, 3)
    )

def gen_int_def(_):
    return int(
        np.random.normal(10, 3)
    )
```

```python
def gen_vision(mental):
    return int(
        np.random.normal(mental, 2.0)
    )

def gen_off_the_ball(mental):
    return int(
        np.random.normal(mental, 2.0)
    )

def gen_positioning(mental):
    return int(
        np.random.normal(mental, 2.0)
    )
```

```python
df_p = pd.DataFrame(
    [
        [gen_height()] for i in range(NUM_PLAYERS)
    ],
    columns=["Height"]
)
df_p["Weight"] = df_p.Height.map(gen_weight)

df_p.Height = df_p.Height.clip(68, 86)

df_p["_technical"] = df_p.Height.map(gen_technical)
df_p["_mental"] = df_p.index.map(gen_mental)

df_p["Agility"] = df_p.Weight.map(gen_agility)  # todo: also include height adjustment for both of these (df.apply)
df_p["Strength"] = df_p.Weight.map(gen_strength)
df_p["VerticalReach"] = df_p.Height.map(gen_vertical)

df_p["BallHandling"] = df_p.apply(gen_ball_handling, axis=1)
df_p["Finishing"] = df_p["_technical"].map(gen_finishing)
df_p["Shooting"] = df_p["_technical"].map(gen_shooting)
df_p["Rebounding"] = df_p.index.map(gen_rebounding)
df_p["PerimeterDefense"] = df_p.index.map(gen_per_def)
df_p["InteriorDefense"] = df_p.index.map(gen_int_def)

df_p["Vision"] = df_p["_mental"].map(gen_vision)
df_p["OffTheBall"] = df_p["_mental"].map(gen_off_the_ball)
df_p["Positioning"] = df_p["_mental"].map(gen_positioning)

for c in df_p.columns:
    if c not in ["Height", "Weight"] and "_" not in c:
        df_p[c] = df_p[c].clip(1, 20)

df_p["ID"] = df_p.index + 1
df_p["CharacterID"] = df_p.index + 1 + NUM_COACHES
df_p["SchoolID"] = ""

births = df.loc[df.PlayerID > 0].Birth
df_p["Eligibility"] = births.map(lambda y: ["FR", "SO", "JR", "SR", "SU"][2025 - y - 19]).values

df_p = df_p.drop(columns=[c for c in df_p.columns if "_" in c])

df_p
```
