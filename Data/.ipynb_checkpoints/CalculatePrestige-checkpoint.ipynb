```python
import pandas as pd
import re
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("all_schools_raw.tsv", sep="\t")
df = df[["School", "Conference", "Tournament appearances", "Final Four appearances", "Championship wins"]]
df = df.loc[df.Conference.isin(["America East", "The American", "ACC", "Big 12", "Horizon League", "Mountain West"])].reset_index(drop=True)
df.index = df.index + 1
df.columns = ["School", "Conference", "Tournaments", "FinalFours", "Championships"]
df
```

```python
re.sub("\D", "", "2007*")
```

```python
test = "(5) 2006, 2007*, 2013, 2014vacated, 2015"

def process_string(s):
    s = s.replace(",", " ")
    l = s.split()[1:]
    l = [int(re.sub("\D", "", y)) for y in l]
    return [year for year in l if year > 1985]

process_string(test)
```

```python
df["Tournaments"] = df["Tournaments"].map(process_string)
df["FinalFours"] = df["FinalFours"].map(process_string)
df["Championships"] = df["Championships"].map(process_string)
```
