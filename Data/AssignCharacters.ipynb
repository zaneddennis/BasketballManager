## Recruit Players

```python
def get_rating(row):
    r_physical = 2 * np.mean([row["Agility"], row["Strength"], row["VerticalReach"]])
    r_technical = 1 * np.mean([row["BallHandling"], row["Finishing"], row["Shooting"], row["Rebounding"], row["PerimeterDefense"], row["InteriorDefense"]])
    r_mental = 1 * np.mean([row["Vision"], row["OffTheBall"], row["Positioning"]])
    return np.sum([r_physical, r_technical, r_mental])

recruits = players[["ID"]].copy()
recruits["Rating"] = players[
    ["Agility", "Strength", "VerticalReach", "BallHandling", "Finishing", "Shooting", "Rebounding", "PerimeterDefense", "InteriorDefense", "Vision", "OffTheBall", "Positioning"]
].apply(get_rating, axis=1)
recruits = recruits.sort_values(by="Rating", ascending=False)
recruits
```

```python
priority = schools[["ID", "PrestigeCurrent"]].copy()
priority["Priority"] = priority["PrestigeCurrent"].map(lambda x: x**3)
priority["Size"] = 0
priority
```

```python
results = []
for i in range(900):
    priority_now = priority.loc[priority.Size < 12]
    
    if len(priority_now) > 0:
        school = priority_now.sample(1, weights=priority_now.Priority)["ID"].iloc[0]
        results.append(school)
        
        priority.loc[priority.ID == school, "Size"] += 1
    else:
        results.append("")

recruits["School"] = results
```

```python
recruits = recruits.sort_values(by="ID")
recruits
```

```python
players["SchoolID"] = recruits.School
players
```

```python
players.to_sql("Players", conn, if_exists="replace", index=False)
```

## Make Teams

```python
teams = pd.DataFrame(
    [
        [i + 1, school_id, 2025, 0, 0] for i, school_id in enumerate(school_ids)
    ], columns=["ID", "SchoolID", "Year", "Wins", "Losses"]
)

teams
```

```python
players_grouped = players.loc[players.SchoolID != "", ["ID", "SchoolID"]].groupby(by="SchoolID")["ID"].apply(lambda l: ",".join(str(p) for p in l))
players_grouped = pd.DataFrame(players_grouped)
players_grouped.columns = ["Players"]
players_grouped
```

```python
teams = pd.merge(teams, coaches.loc[coaches.SchoolID != "", ["ID", "SchoolID"]], on="SchoolID", suffixes=("", "_Coach"))
teams = pd.merge(teams, players_grouped, left_on="SchoolID", right_index=True)
teams = teams.rename(columns={"ID_Coach": "HeadCoach"})
teams
```

```python
teams.to_sql("Teams", conn, if_exists="append", index=False)
```
